```python
# Dwie różne liczby całkowite a i b większe od 1 nazwiemy skojarzonymi,
# jeśli suma wszystkich różnych dodatnich dzielników a mniejszych od a jest równa
# b+1, a suma wszystkich różnych dodatnich dzielników b mniejszych od b jest
# równa a+1. Ułożyć i zapisać w wybranej przez siebie notacji (kod w Python
# albo pseudokod albo lista kroków) algorytm, który znajdzie i wypisze liczęb
# skojarzoną z daną liczbą całkowitą (większą od 1) a lub komunikat NIE, jeśli
# taka liczba nie istnieje. (10 pkt)
```

```python
def aib(a,b):
    dza=[]
    dzb=[]
    for i in range(1, a + 1):
        if a % i == 0 and i < a:
            dza.append(i)
        if b % i == 0 and i < b:
            dzb.append(i)
    sumaA=0
    sumaB=0
    for i in dza:
        sumaA += i
    for i in dzb:
        sumaB += i
#     print(dza,dzb,sumaA,sumaB)
    if sumaA == b+1 and sumaB == a+1:
        return f"Liczba skojarzona z {a} to {b}"
    else:
        return "NIE"
    return (dza,dzb,sumaA,sumaB)

#TEST
for i in range(40,50):
    for j in range(70,80):
        print(aib(i,j))
```
