Zaimplementuj klasę wektor:
(dodawanie, mnożenie przez skalar, odejmowanie, wyznaczenie długości, iloczyn skalarny wektorów, czy wektory sa równe, czy wektory są prostopadłe, czy wektory są równoległe)

```python
import math
import numbers

class Vector:
  def __init__(self, vector):
    self.vector = vector

  def sprawdzenie(self, vector2):
    if len(self.vector) == len(vector2.vector):
      return True
    else:
      raise ValueError('Wektory znajdują się w innych wymiarach.')


  def dodaj(self, vector2):
    self.sprawdzenie(vector2)
    for el in range(len(self.vector)):
      self.vector[el] += vector2.vector[el]
    return self

  def odejmij(self, vector2):
    self.sprawdzenie(vector2)
    for el in range(len(self.vector)):
      self.vector[el] -= vector2.vector[el]
    return self

  def mnozenie(self, skalar):
      for el in range(len(self.vector)):
        self.vector[el] *= skalar
      return self

  def dlugosc(self):
    wynik = 0
    for el in range(len(self.vector)):
      wynik += self.vector[el]**2
    return math.sqrt(wynik)

  def iloczyn_skalarny(self, vector2):
    self.sprawdzenie(vector2)
    wynik = 0
    wynik2 = 0
    for el in range(len(self.vector)):
      wynik = self.vector[el] * vector2.vector[el]
      wynik2 += wynik
    return wynik2

  def rowne(self, vector2):
    self.sprawdzenie(vector2)
    dlugosc = 0
    for el in range(len(self.vector)):
      if self.vector[el] == vector2.vector[el]:
        dlugosc +=1
    if dlugosc == len(self.vector):
      return 'Wektory sa sobie równe.'
    else:
      return 'Wektory NIE sa sobie równe.'

  def prostopadle(self, vector2):
    self.sprawdzenie(vector2)
    if self.iloczyn_skalarny(vector2) == 0:
      return 'Wektory są prostopadłe.'
    else:
      return 'Wektory NIE są prostopadłe.'

  def rownolegle(self,vector2):
    self.sprawdzenie(vector2)
    k = 0
    k1 = 0
    sl = 'Wektor jest równoległy'
    ls = 'Wektor NIE jest równoległy'
    for el in range(len(self.vector)):
      if vector2.vector[el] == 0 and self.vector[el] == 0:
        return  sl
      elif vector2.vector[el] == 0 and vector2.vector[el] != self.vector[el]:
        return ls
      k = self.vector[el] / vector2.vector[el]
      k1 += k
    if k1 / (len(self.vector)) == k:
      return sl
    else:
      return ls

  def __str__(self):
        wypisz = ""
        for el in range(len(self.vector)):
            if type(self.vector[el]) == int:
                if el == len(self.vector) - 1:
                    wypisz += str(self.vector[el])
                else:
                    wypisz += str(self.vector[el]) + ", "
        return wypisz
```
